```python
dataset["labels"] = dataset["class"].map({0: "Offensive Language or Hate Speech",
                                         1: "Offensive Language or Hate Speech",
                                         2: "No hate or offensive Language"})
```

```python
data = dataset[["tweet", "labels"]]
```

import re
import nltk
import string

```python
stopwords = set(nltk.corpus.stopwords.words("english"))
```

```python
stopwords
```

```python
stemmer = nltk.SnowballStemmer("english")
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
def data_clean(text):
    text = str(text).lower()
    text = re.sub("hhtps?://S+www\\.S+", "", text)
    text = re.sub("\\[.*?\\]", "", text)
    text = re.sub("<.*?>+", "", text)
    text = re.sub("[%s]" % re.escape(string.punctuation), "", text)
    text = re.sub("\\n", "", text)
    text = re.sub("\\w*\\d\\w*", "", text)
    words = [stemmer.stem(word) for word in text.split() if word not in stopwords]
    text = " ".join(words)
    return text
```

```python
data["tweet"] = data["tweet"].apply(data_clean)
```

```python
data
```

```python
X = np.array(data["tweet"])
Y = np.array(data["labels"])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv = CountVectorizer()
X = cv.fit_transform(X)
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth = 100)
dt.fit(X_train, Y_train)
```

```python
Y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = confusion_matrix(Y_test, Y_pred)
cm
```

```python
print(classification_report(Y_test, Y_pred))
```
